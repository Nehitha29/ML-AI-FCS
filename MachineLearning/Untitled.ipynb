```python
# Considering iris dataset, design a Bayesian classifier to classify the
# test data. Divide the dataset into training and testing data into (a) 70:30 ratio (b) 80:20
# ratio. Calculate the accuracy and error rate
```

```python
# using bayes classifier , calculating likelyhood and priors
```

```python
from sklearn import datasets
import  pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names= iris.target_names
df = pd.DataFrame(X , columns= feature_names)
df['target']= y
```

```python
df
```

```python
def bayesian_classifier(train_data, test_data, a=1, b=1):

    X_train = train_data.drop('target', axis=1).values
    y_train = train_data['target'].values
    X_test = test_data.drop('target', axis=1).values
    y_test = test_data['target'].values

    classes, counts = np.unique(y_train, return_counts=True)
    priors = counts / len(y_train)

    likelihoods = {}
    for i, class_label in enumerate(classes):
        class_data = X_train[y_train == class_label]
        means = np.mean(class_data, axis=0)
        variances = np.var(class_data, axis=0)
        likelihoods[class_label] = (means, variances)
    def predict(x):
        posteriors = []
        for i, class_label in enumerate(classes):
            mean, variance = likelihoods[class_label]
            prior = priors[i]
            likelihood = np.prod(1 / np.sqrt(2 * np.pi * variance) * 
                                 np.exp(- (x - mean)**2 / (2 * variance)), axis=0)
            posterior = prior * likelihood
            posteriors.append(posterior)
        return classes[np.argmax(posteriors)]
    predictions = np.array([predict(x) for x in X_test])
    accuracy = np.mean(predictions == y_test)
    error_rate = 1 - accuracy
    return accuracy, error_rate
    
```

```python
def evaluate_classifier(split_ratio):

    train_data, test_data = train_test_split(df, test_size=split_ratio, random_state=42)

    accuracy, error_rate = bayesian_classifier(train_data, test_data)
    return accuracy, error_rate
```

```python
accuracy_70_30, error_rate_70_30 = evaluate_classifier(0.30)
 

accuracy_80_20, error_rate_80_20 = evaluate_classifier(0.20)
 

print(f"70:30 Split - Accuracy: {accuracy_70_30:.2f}, Error Rate: {error_rate_70_30:.2f}")
print(f"80:20 Split - Accuracy: {accuracy_80_20:.2f}, Error Rate: {error_rate_80_20:.2f}")
```
